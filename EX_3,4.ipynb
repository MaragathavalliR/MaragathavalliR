```python
import pandas as pd
import math

class Node:
    def __init__(self, feature=None, value=None, leaf=False, label=None):
        self.feature = feature  # Feature to split on
        self.value = value      # Value of the feature to split on
        self.leaf = leaf        # True if it's a leaf node
        self.label = label      # Class label for leaf nodes
        self.children = {}      # Dictionary of child nodes

def entropy(samples):
    counts = samples.value_counts()
    total_samples = len(samples)
    entropy_value = 0

    for count in counts:
        p = count / total_samples
        entropy_value -= p * math.log2(p)

    return entropy_value

def information_gain(data, split_feature, target_feature):
    total_entropy = entropy(data[target_feature])
    feature_values = data[split_feature].unique()

    weighted_entropy = 0
    for value in feature_values:
        subset = data[data[split_feature] == value]
        subset_entropy = entropy(subset[target_feature])
        weighted_entropy += (len(subset) / len(data)) * subset_entropy

    return total_entropy - weighted_entropy

def build_tree(data, features, target_feature):
    if len(data[target_feature].unique()) == 1:
        return Node(leaf=True, label=data[target_feature].iloc[0])

    if len(features) == 0:
        dominant_class = data[target_feature].value_counts().idxmax()
        return Node(leaf=True, label=dominant_class)

    best_feature = max(features, key=lambda f: information_gain(data, f, target_feature))
    root = Node(feature=best_feature)

    for value in data[best_feature].unique():
        subset = data[data[best_feature] == value]
        if len(subset) == 0:
            dominant_class = data[target_feature].value_counts().idxmax()
            root.children[value] = Node(leaf=True, label=dominant_class)
        else:
            root.children[value] = build_tree(subset, [f for f in features if f != best_feature], target_feature)

    return root

def print_tree(node, indent="", coverage=None):
    if node.leaf:
        print(indent + "Leaf: " + node.label)
    else:
        print(indent + "Attribute: " + node.feature)
        print(indent + "Values(" + node.feature + "):", data[node.feature].unique())
        print(indent + "S = [{}+, {}-] = {:.2f}".format(len(data[data[target_feature] == "yes"]), len(data[data[target_feature] == "no"]), entropy(data[target_feature])))
        for value in data[node.feature].unique():
            subset = data[data[node.feature] == value]
            subset_entropy = entropy(subset[target_feature])
            print(indent + "S{} = [{}+, {}-] = {:.2f}".format(value, len(subset[subset[target_feature] == "yes"]), len(subset[subset[target_feature] == "no"]), subset_entropy))
        print(indent + "Expected Information for " + node.feature + ":")
        info_before = entropy(data[target_feature])
        info_after = 0
        for value in data[node.feature].unique():
            subset = data[data[node.feature] == value]
            info_after += entropy(subset[target_feature]) * len(subset) / len(data)
        print(indent + "info({}) = {:.3f} bits".format(data[target_feature].value_counts().values.tolist(), info_after))
        print(indent + "Information gain:")
        print(indent + "gain({}, {}) = {:.3f} bits".format(node.feature, value, info_before - info_after))
        print()
        for value, child in node.children.items():
            print(indent + "Value " + value + ":")
            print_tree(child, indent + "  ")

data = pd.read_csv("humidity.csv")  # Replace with the path to your dataset file
target_feature = "Play"
features = data.columns[:-1]

print("Total coverage count: {}+".format(len(data[data[target_feature] == "yes"])),
      "{}-".format(len(data[data[target_feature] == "no"])))
print("Total entropy:", entropy(data[target_feature]))
print()

root_node = build_tree(data, features, target_feature)
print_tree(root_node)

# Find the head node identification
max_info_gain = 0
head_node = None
for feature in features:
    info_gain = information_gain(data, feature, target_feature)
    if info_gain > max_info_gain:
        max_info_gain = info_gain
        head_node = feature

print("\nThe root node should be identified based on the maximum information gain.")
print("In this case, the attribute with the maximum information gain is '{}' with a gain of {:.3f} bits.".format(head_node, max_info_gain))

def tree_to_dict(node):
    if node.leaf:
        return node.label
    else:
        tree_dict = {}
        for value, child in node.children.items():
            tree_dict[value] = tree_to_dict(child)
        return tree_dict

decision_tree_structure = tree_to_dict(root_node)
print("Decision Tree Structure:")
print(decision_tree_structure)
```
